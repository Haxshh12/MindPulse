```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
import pickle

# Load data
data = pd.read_csv('C:\\Users\\Harsh Jethwa\\OneDrive\\Desktop\\the_new_mindpulse\\data\\user_inputs.csv')

# Drop columns that are not features
X = data.drop(columns=['email', 'name', 'timestamp'])

# Label (for now dummy)
y = np.random.choice(['Low', 'Moderate', 'High'], size=len(X))

# Define categorical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

# FIXED categories for each column to avoid future unseen category issues
fixed_categories = {
    'exercise': ['Daily', 'Weekly', 'Occasional', 'Never'],
    'sleep_quality': ['Poor', 'Average', 'Good'],
    'diet': ['Unhealthy', 'Average', 'Healthy'],
    # add all other categorical columns here
}

# OneHotEncoder with fixed categories
encoder = OneHotEncoder(categories=[fixed_categories[col] for col in categorical_cols],
                        handle_unknown='ignore', sparse=False)

X_encoded = encoder.fit_transform(X[categorical_cols])
X_encoded = pd.DataFrame(X_encoded, columns=encoder.get_feature_names_out(categorical_cols))

# Keep numeric columns as they are
numeric_cols = X.drop(columns=categorical_cols)
X_final = pd.concat([numeric_cols.reset_index(drop=True), X_encoded.reset_index(drop=True)], axis=1)

# Train model
model = RandomForestClassifier()
model.fit(X_final, y)

# Save both model and encoder
with open("ml_model/mental_health_model.pkl", "wb") as f:
    pickle.dump((model, encoder, numeric_cols.columns.tolist()), f)

print("Model trained with fixed schema.")
```
